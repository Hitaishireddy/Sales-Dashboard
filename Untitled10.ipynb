```python

import pandas as pd
import matplotlib.pyplot as plt


df=pd.read_csv('/content/Sales Latest - Sheet1.csv')

model_summary = df.groupby('Model').agg(
    Total_Quantity_Sold=('Quantity Sold', 'sum'),
    Total_Profit=('Profit', 'sum')
).reset_index()

# Create the plot
plt.figure(figsize=(12, 6))

plt.plot(model_summary['Model'], model_summary['Total_Quantity_Sold'], marker='o', label='Quantity Sold')
plt.plot(model_summary['Model'], model_summary['Total_Profit'], marker='s', label='Profit')

plt.title('Quantity Sold & Profit by Model')
plt.xlabel('Model')
plt.ylabel('Amount in $')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
