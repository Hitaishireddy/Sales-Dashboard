```python

import pandas as pd
import matplotlib.pyplot as plt


df=pd.read_csv('/content/Sales Latest - Sheet1.csv')
null_counts = df.isnull().sum()
print("Null values per column:")
print(null_counts)

model_summary = df.groupby('Model').agg(
    Total_Quantity_Sold=('Quantity Sold', 'sum'),
    Total_Profit=('Profit', 'sum')
).reset_index()


import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

fig, ax1 = plt.subplots(figsize=(12, 6))

ax1.plot(model_summary['Model'], model_summary['Total_Quantity_Sold'],
         marker='o', color='b', label='Quantity Sold')
ax1.set_xlabel('Model')
ax1.set_ylabel('Quantity Sold', color='b')
ax1.tick_params(axis='y', labelcolor='b')

ax2 = ax1.twinx()
ax2.plot(model_summary['Model'], model_summary['Total_Profit'],
         marker='s', color='orange', label='Profit')
ax2.set_ylabel('Profit ($)', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

plt.title('Quantity Sold & Profit by Model')
fig.tight_layout()
plt.grid(True)

plt.show()
```
